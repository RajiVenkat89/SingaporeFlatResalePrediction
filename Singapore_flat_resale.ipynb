```python
def plot_data(df,col):
    plt.figure(figsize=(25,5))
    plt.subplot(1,3,1)
    sns.boxplot(df[col])
    plt.title("Boxplot of "+str(col))
    plt.subplot(1,3,2)
    sns.histplot(data=df,x=col,kde=True,bins=50)
    plt.title("Distribution of "+str(col))
    plt.subplot(1,3,3)
    sns.violinplot(df[col])
    plt.title("ViolinPlot of "+str(col))
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score
X=df.drop(['town', 'flat_type', 'block', 'street_name', 'storey_range','flat_model','block_Alphabet','resale_price'],axis=1)
y=df['resale_price']

x_train, x_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
mae_train={'LinearRegression':[],'KNeighborsRegressor':[],'GradientBoostingRegressor':[],'RandomForestRegressor':[],'DecisionTreeRegressor':[]}
mae_test={'LinearRegression':[],'KNeighborsRegressor':[],'GradientBoostingRegressor':[],'RandomForestRegressor':[],'DecisionTreeRegressor':[]}
mse_train={'LinearRegression':[],'KNeighborsRegressor':[],'GradientBoostingRegressor':[],'RandomForestRegressor':[],'DecisionTreeRegressor':[]}
mse_test={'LinearRegression':[],'KNeighborsRegressor':[],'GradientBoostingRegressor':[],'RandomForestRegressor':[],'DecisionTreeRegressor':[]}
rmse_train={'LinearRegression':[],'KNeighborsRegressor':[],'GradientBoostingRegressor':[],'RandomForestRegressor':[],'DecisionTreeRegressor':[]}
rmse_test={'LinearRegression':[],'KNeighborsRegressor':[],'GradientBoostingRegressor':[],'RandomForestRegressor':[],'DecisionTreeRegressor':[]}
r2_train={'LinearRegression':[],'KNeighborsRegressor':[],'GradientBoostingRegressor':[],'RandomForestRegressor':[],'DecisionTreeRegressor':[]}
r2_test={'LinearRegression':[],'KNeighborsRegressor':[],'GradientBoostingRegressor':[],'RandomForestRegressor':[],'DecisionTreeRegressor':[]}
models=[LinearRegression(),KNeighborsRegressor(),GradientBoostingRegressor(),RandomForestRegressor(max_depth=10),DecisionTreeRegressor()]
for i in range(1,6):
    for model in models:
        model.fit(x_train,y_train)
        train_pred=model.predict(x_train)
        test_pred=model.predict(x_test)
        print(f"****{type(model).__name__}****")
        mae_train[type(model).__name__].append(mean_absolute_error(y_train,train_pred))
        mae_test[type(model).__name__].append(mean_absolute_error(y_test,test_pred))
        print(f"MAE_Train: {mean_absolute_error(y_train,train_pred)}")
        print(f"MAE_Test: {mean_absolute_error(y_test,test_pred)}")
        mse_train[type(model).__name__].append(mean_squared_error(y_train,train_pred))
        mse_test[type(model).__name__].append(mean_squared_error(y_test,test_pred))
        print(f"MSE_Train: {mean_squared_error(y_train,train_pred)}")
        print(f"MSE_Test: {mean_squared_error(y_test,test_pred)}")
        rmse_train[type(model).__name__].append(root_mean_squared_error(y_train,train_pred))
        rmse_test[type(model).__name__].append(root_mean_squared_error(y_test,test_pred))
        print(f"RMSE_Train: {root_mean_squared_error(y_train,train_pred)}")
        print(f"RMSE_Test: {root_mean_squared_error(y_test,test_pred)}")
        r2_train[type(model).__name__].append(r2_score(y_train,train_pred))
        r2_test[type(model).__name__].append(r2_score(y_test,test_pred))
        print(f"R2Score_Train: {r2_score(y_train,train_pred)}")
        print(f"R2Score_Test: {r2_score(y_test,test_pred)}")
```

```python
df1=pd.DataFrame(r2_train)
df2=pd.DataFrame(r2_test)
df1['class']='Train'
df2['class']='Test'
df_analyse=pd.concat([df1,df2])
op=pd.melt(df_analyse,id_vars=['class'],value_vars=['LinearRegression','KNeighborsRegressor','GradientBoostingRegressor','RandomForestRegressor','DecisionTreeRegressor'])
sns.scatterplot(data=op,x='variable',y='value',hue='class')
plt.xticks(rotation=60)
plt.show()
print(df_analyse)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score
X=df.drop(['town', 'flat_type', 'block', 'street_name', 'storey_range','flat_model','block_Alphabet','resale_price'],axis=1)
y=df['resale_price']

x_train, x_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
model=DecisionTreeRegressor(max_depth=50, min_samples_split=5,min_samples_leaf=6,max_features=None)
model.fit(x_train,y_train)
train_pred=model.predict(x_train)
test_pred=model.predict(x_test)
print(f"MAE_Train: {mean_absolute_error(y_train,train_pred)}")
print(f"MAE_Test: {mean_absolute_error(y_test,test_pred)}")
print(f"MSE_Train: {mean_squared_error(y_train,train_pred)}")
print(f"MSE_Test: {mean_squared_error(y_test,test_pred)}")
print(f"RMSE_Train: {root_mean_squared_error(y_train,train_pred)}")
print(f"RMSE_Test: {root_mean_squared_error(y_test,test_pred)}")
print(f"R2Score_Train: {r2_score(y_train,train_pred)}")
print(f"R2Score_Test: {r2_score(y_test,test_pred)}")
model.feature_importances_
```
